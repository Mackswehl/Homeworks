```
class Node():
  
  def __init__(self,data):
    self.data = data
    self.leftChild = None
    self.rightChild = None
  
  def value(self):
    return self.data
  
  
class BinTree():
    
  def __init__(self,A):
    self.size = 0
    self.root = None
    for i in A:
      self.insert(i)
      
  def insert(self,v,curRoot = 0):
    newNode = Node(v)
    
    if curRoot == 0:
      curRoot = self.root
      
    if self.size == 0:
      self.root = newNode
      self.size += 1
    elif v < curRoot.value():
      if curRoot.leftChild == None:
        curRoot.leftChild = newNode
        self.size += 1
      else:
        self.insert(v,curRoot.leftChild)
    else:
      if curRoot.rightChild == None:
        curRoot.rightChild = newNode
        self.size += 1
      else:
        self.insert(v,curRoot.rightChild)
        
  def has(self,v,curRoot = 0):
    if curRoot == 0:
      curRoot = self.root
    if curRoot == None:
      return False
    elif curRoot.value() == v:
      return True
    elif v < curRoot.value():
      return self.has(v,curRoot.leftChild)
    else:
      return self.has(v,curRoot.rightChild)
      
  def get_ordered_list(self):
    return self.inOrder(self.root)
      
  def inOrder(self,curRoot,retList = None):
    if retList == None:
      retList = []
    if (curRoot != None):
      self.inOrder(curRoot.leftChild,retList)
      retList.append(curRoot.value())
      self.inOrder(curRoot.rightChild,retList)
    return retList
  
  def clear(self):
    self.postOrderClear(self.root)
    self.root = None
    self.size = 0
  
  def postOrderClear(self,curRoot):
      if curRoot.leftChild != None:
        self.postOrderClear(curRoot.leftChild)
        curRoot.leftChild = None
      if curRoot.rightChild != None:
        self.postOrderClear(curRoot.rightChild)
        curRoot.rightChild = None
        
        
          
    
    
    
    
    
  
  
    
```
